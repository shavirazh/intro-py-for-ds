```python
import matplotlib.pyplot as plt
```

```python
afghanistanImigran = imigran_by_country.loc['Afghanistan', range(1980,2013)]
afghanistanImigran
```

```python
years = range(1980,2013)
chn_ind_imigrant = imigran_by_country.loc[['China','India'], years]
```
